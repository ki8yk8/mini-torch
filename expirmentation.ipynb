```python
import matplotlib.pyplot as plt
import numpy as np
```

# Numeric and Manual Derivative

```python
f = lambda x: x*x
```

```python
x = np.arange(-10, 10, 1)
y = [f(i) for i in x]

# plotting the result
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
def f_derivative_manual(x):
	return 2*x
```

```python
print(f_derivative_manual(2.5))
print(f_derivative_manual(-2.5))
print(f_derivative_manual(0))    # notice the rounding off error because at 0 the actual derivative is also zero
```

```python
# computing the difference at any point of time using numeric differeniation method
# using the first principle of derivative
def f_derivative_numeric(x, h=1e-4):
	return (f(x+h) - f(x))/h
```

```python
print(f_derivative_numeric(2.5))
print(f_derivative_numeric(-2.5))
print(f_derivative_numeric(0))    # notice the rounding off error because at 0 the actual derivative is also zero
```

```python
# plotting the error difference of both the methods
y_derivative_manual = [f_derivative_manual(i) for i in x]
y_derivative_numeric = [f_derivative_numeric(i) for i in x]

plt.plot(x, y, label="function")
plt.plot(x, y_derivative_manual, label="manual derivative")
plt.plot(x, y_derivative_numeric, label="numeric approximation")
plt.legend()
plt.grid()
plt.show()

# notice they are similar but has rounding off and truncation error which can propagate if the chain is long
```

## Partial derivative 

```python
a, b = 3, 4
h = 1e-4

f = lambda a, b: a*a+2*a*b+b*b
print(f(a, b))
```

```python
# partial derivative with respect to a means we increase the value of a and b is constant
d_f_wrt_a = (f(a+h, b)-f(a, b))/h
print(d_f_wrt_a)    # manually = 2a+2b+0 = 2*3+2*4 = 14
```

```python
# with respect to b equals
print((f(a, b+h)-f(a, b))/h)    # manually = 0+2a+2b = 14
```

# Creating AutoGrad

```python
class Value:
	def __init__(self, data):
		self.data = data

	def __mul__(self, other):
		result = self.data * other.data
		return Value(data=result)
	
	def __add__(self, other):
		result = self.data + other.data
		return Value(data=result)
	
	def __repr__(self):
		print(f"Value({self.data:.4f})")
```
